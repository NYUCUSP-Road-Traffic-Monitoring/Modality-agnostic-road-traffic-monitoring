```python
import pandas as pd
```

```python
anno = pd.read_csv("annotations_by_frame.csv")
anno = anno.fillna(0)

anno.head()
```

```python
anno.shape
```

```python
def get_video_name(name):
    name = name.split("data/task_")[1]
    name = name.split("-2021")[0]
    name = name.split(".mp4")[0]
    return name + ".mp4"
```

```python
anno["video_name"] = anno["file_name"].apply(get_video_name)
```

```python
anno[:20]
```

```python
bosch=anno[~anno['video_name'].str.contains('street')]
```
